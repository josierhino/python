```python
import yfinance as yf

nflx = yf.Ticker('NFLX')
```

```python
nflx.info
```

Above we can see information about the company Netflix. Below we will find Netflix's income statement as reported on Yahoo Finance

```python
nflx.quarterly_income_stmt
```

 Let's inspect the information. Netflix implemented the Password Sharing policy in July 2023. It was announced in May 2023. Let's take a look at revenue before and after the Password Sharing policy. We will use 2022-12-31 as our Before benchmark and 2023-06-30 as our After benchmark.

```python
# Clean up output to only show Total Revenue (y axis) and 2022-12-31, 2023-03-31, and 2023-06-30 (x axis)

quarters_cleaned = nflx.quarterly_income_stmt[['2022-12-31', '2023-03-31', '2023-06-30']]


quarters_cleaned
```

```python
clean = quarters_cleaned.loc[['Total Revenue']]
clean
```

Let's examine our data now. It looks like Netflix's revenue did increase since the password sharing policy!

```python
import matplotlib.pyplot as plt
```

Let's plot our cleaned data now! 

```python
clean.T.plot()
```

Looking above, we can see that Q4 represents 2022-12-31 and Q1 represents 2022-03-31 and Q2 represents 2022-06-30, respectively. We can see that revenue increased significantly then stabled.
